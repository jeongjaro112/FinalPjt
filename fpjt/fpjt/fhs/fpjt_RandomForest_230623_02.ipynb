```python
import pandas as pd
import re
import numpy as np
from sklearn.model_selection import train_test_split
import statsmodels.api as sm

# 데이터프레임 불러오기
stock_select = pd.read_csv('./stockInfo_20230609-100747.csv', encoding='utf-8', thousands=',')

# 정규식 패턴
pattern = re.compile(r'[\d,]+\.?\d*')

# 모든 칼럼에 적용하는 함수
def clean_cell(cell):
    cell = re.sub(r'[\r\n\t\s]', '', str(cell))  # 이스케이프 시퀀스 및 공백 제거
    if cell == '':
        return None
    numeric_value = pattern.findall(cell)  # 숫자 값 추출
    return numeric_value[0] if numeric_value else cell

# 모든 칼럼에 함수 적용
stock_select = stock_select.applymap(clean_cell)

# 결측값 처리
stock_select = stock_select.fillna(0)  # 결측값을 0으로 대체하거나

# 유형 변환
stock_select['투자의견'] = pd.to_numeric(stock_select['투자의견'], errors='coerce')

# 'select' 컬럼 초기화
stock_select['select'] = 0

# 'select' 값 설정
stock_select.loc[stock_select['투자의견'] > 3.5, 'select'] = 1

# 모델 적합을 위한 데이터 준비
X = stock_select[['EV_EBITDA', 'PBR', 'PER', 'ROE', '배당수익율', '부채비율']]
y = stock_select['select']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)

# 데이터 유형 변환
X_train = X_train.replace('-', np.nan)
X_train = X_train.replace(',', '', regex=True).replace('inf', np.nan).astype(float)
X_train = X_train.fillna(0)

X_test = X_test.replace('-', np.nan)
X_test = X_test.replace(',', '', regex=True).replace('inf', np.nan).astype(float)
X_test = X_test.fillna(0)

# 모델 적합
X_train = sm.add_constant(X_train)  # 상수항 추가
linear_model = sm.OLS(y_train, X_train)  # 선형 회귀 모델
result = linear_model.fit()  # 모델 적합

# 결과 출력
print(result.summary())
```

```python
# R-squared(결정 계수): 이 값은 모델이 종속 변수의 변동성을 얼마나 잘 설명하는지 나타냄. 
# R-squared 값은 0과 1 사이의 범위에서 변동하며, 1에 가까울수록 모델이 종속 변수의 변동성을 잘 설명. 
# 위의 경우, R-squared 값은 0.038로 매우 낮다. 즉, 독립 변수들이 종속 변수의 변동성을 매우 약하게 
# 설명하고 있다.

# Adj. R-squared(조정 결정 계수): R-squared 값을 보정한 값으로, 독립 변수의 수와 샘플의 크기를 
# 고려하여 모델의 설명력을 보다 정확하게 평가함. Adj. R-squared 값도 0과 1 사이의 범위에서 변동하며, \
# 위의 경우 0.034로 매우 낮다.

# F-statistic: 회귀 모델의 유의성을 평가하는 값이다. F-statistic 값이 클수록 모델의 유의성이 높다. 
# 위의 경우 F-statistic 값은 11.25로, 유의성이 있는 모델이다. Prob (F-statistic) 값은 F-statistic에 대한 p-value를 나타내며, 
# 이 값이 매우 작은 것을 보아 유의미한 모델이라고 할 수 있다.

# 회귀 계수(coef): 각 독립 변수의 회귀 계수 값과 해당 계수의 통계적 유의성에 대한 정보가 제공된다. 
# 회귀 계수는 해당 독립 변수가 종속 변수에 미치는 영향을 나타내며, 계수가 양수이면 해당 독립 변수가 
# 종속 변수를 증가시키는 영향을 가지고, 음수이면 감소시키는 영향을 가진다고 해석할 수 있습니다. 

# P>|t| 값은 회귀 계수의 통계적 유의성에 대한 p-value를 나타내며, 일반적으로 0.05보다 작으면 통계적으로 유의미한 계수다.

# Omnibus, Prob(Omnibus), Jarque-Bera (JB), Skew, Kurtosis: 이는 회귀 모델의 잔차(residual)에 대한 
# 정규성과 대칭성에 대한 검정 결과를 제공한다. 이 값들은 회귀 모델의 가정을 평가하는 데 사용됩니다.

# Durbin-Watson: 이 값은 회귀 모델의 잔차 사이에 자기 상관(autocorrelation)이 존재하는지를 평가합니다. 
# 값이 2에 가까울수록 자기 상관이 존재하지 않는다는 가정을 만족한다.

# Cond. No.: 이 값은 다중공선성(multicollinearity)의 정도를 나타내는 조건수(condition number)입니다. 
# 조건수가 큰 경우 다중공선성이 존재할 수 있으며, 이는 독립 변수들 사이에 강한 상관관계가 있음을 의미.

# 위의 결과를 종합하면, 주어진 독립 변수들이 종속 변수를 설명하는 데 매우 약한 영향을 가지고 있으며, 
# 모델의 설명력은 매우 낮다. 또한, 회귀계수 중 일부는 통계적으로 유의미한 것으로 나타나지만, 다른 독립변수들의 
# 회귀계수는 유의미하지 않다. 유의미한 것으로 나타난 독립 변수는 'PBR', '배당수익율' 및 '부채비율'이다. 
# 이 독립 변수들은 p-value가 0.05보다 작으므로, 통계적으로 유의미한 영향을 가지고 있다고 할 수 있다. 
# 다른 독립 변수들인 'EV_EBITDA', 'PER', 'ROE'의 경우에는 p-value가 0.05보다 크므로, 
# 통계적으로 유의미한 영향을 가지지 않는 것으로 판단해야 한다.
# 모델의 잔차에 대한 정규성과 대칭성 검정에서는 이상이 없는 것으로 
# 나타났고, 자기 상관이 없는 것으로 판단됨. 
# 결론적으로, 주어진 선형 회귀 모델은 주가수익율을 설명하는 데 유의미한 결과를 도출하지 못했으며, 
# 다른 모델이나 독립 변수들의 조합을 고려해볼 필요가 있다.
```

```python
import pandas as pd
import os
import datetime
import glob

file_path = os.path.join(os.getcwd())
print(f'\n☞ 현재의경로: {file_path}\n')
file_list = glob.glob(os.path.join(file_path, 'stockInfo_all_*.csv'))
print(f'\n☞ file_list: {file_list}\n')
file_latest = max(file_list, key = os.path.getctime)
print(f'\n☞ file_latest: {file_latest}\n')
df_sti = pd.read_csv(file_latest)
df_sti

df_sti['PER'] = pd.to_numeric(df_sti['PER'], errors='coerce')
df_sti['PBR'] = pd.to_numeric(df_sti['PBR'], errors='coerce')
df_sti['배당수익율'] = pd.to_numeric(df_sti['배당수익율'], errors='coerce')
df_sti['EV_EBITDA'] = pd.to_numeric(df_sti['EV_EBITDA'], errors='coerce')
df_sti['부채비율'] = pd.to_numeric(df_sti['부채비율'], errors='coerce')
df_sti['ROE'] = pd.to_numeric(df_sti['ROE'], errors='coerce')
df_sti['투자의견'] = pd.to_numeric(df_sti['투자의견'],   errors='coerce')

condi_1 = (df_sti['PER'] <= 20) & (df_sti['PBR'] <= 2) & (df_sti['배당수익율'] >= 2) & (df_sti['EV_EBITDA'] < 5) & (df_sti['부채비율'] <= 100) & (df_sti['ROE'] >= 10)
df_filtered_1 = df_sti.loc[condi_1]
print(df_filtered_1)
#df_filtered_1

condi_2 = df_sti['투자의견'].notna()
df_filtered_2 = df_sti.loc[condi_2]
#df_filtered_2
print(df_filtered_2)

condi_3 = condi_1 | condi_2
df_filtered_3 = df_sti.loc[condi_3]
print(df_filtered_3)

df_filtered_1.to_csv('filtered_1.csv', index=False)
df_filtered_2.to_csv('filtered_2.csv', index=False)
df_filtered_3.to_csv('filtered_3.csv', index=False)
```
